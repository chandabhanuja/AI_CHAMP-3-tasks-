```
list_text = []
import glob, os
os.chdir("/content")
for file in glob.glob("/content/*.pdf"):
  parsedPDF = parser.from_file(file)
  text = parsedPDF['content']
  text = text.replace('\n','')
  list_text.append(text)
  print(file)
```
